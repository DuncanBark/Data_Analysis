```python
from numpy import *
from matplotlib.pyplot import *
%matplotlib inline
%load_ext autoreload
%autoreload 2
poly = np.polynomial.Polynomial

from astropy.io import ascii

from scipy.integrate import quad
from scipy.integrate import trapz
from scipy.integrate import odeint
from scipy.interpolate import interp1d
from scipy.ndimage.filters import gaussian_filter1d
from scipy.misc import derivative
from scipy.optimize import curve_fit
```

```python
sl = ascii.read("Stratologger CF Huntsville 2019.txt", data_start = 16)
em = ascii.read("Easy Mini SLI Export Flight 2019 Huntsville.csv", data_start = 30)
```

```python
sl
# em
```

```python
mtoft = 3.2808

emAltM = em["altitude"]
emPress = em["pressure"]
emAccelM = em["acceleration"]
emVelM = em["baro_speed"]
emTime = em["time"]

emAlt = [x*mtoft for x in emAltM]
emAlt = [x-emAlt[-1] for x in emAlt]
emVel = [x*mtoft for x in emVelM]
emAccel = [x*mtoft for x in emAccelM]

slAlt = sl["col2"]
slVel = sl["col3"]
slTime = sl["col1"]
```

```python
figure(figsize = (16, 15))

subplot(321)
plot(emTime, emAlt)

subplot(322)
plot(slTime, slAlt)

subplot(323)
plot(emTime, emVel)

subplot(324)
plot(slTime, slVel)

subplot(325)
plot(emTime, emAccel)

subplot(326)
# plot(slTime, slAccel)

show()
```

```python
# "smooth" altitude by determining difference between adjacent values and changing it if its within a certain value
# called numerically smoothed values in future references
        
emAltA = [x for x in emAlt] #re-make emAltA so you dont compound adjustments
for x in range(1, len(emAltA)):
    if abs(emAltA[x] - emAltA[x-1]) > 100: #40 is minimum difference to maintain accuracy and relative "smoothness"
        emAltA[x] = emAltA[x-1]
        
slAltA = [x for x in slAlt] #re-make emAltA\ so you dont compound adjustments
for x in range(1, len(slAltA)):
    if abs(slAltA[x] - slAltA[x-1]) > 100: #40 is minimum difference to maintain accuracy and relative "smoothness"
        slAltA[x] = slAltA[x-1]
```

```python
figure(figsize = (15, 5))

subplot(121)
plot(emTime, emAltA)
# xlim(105, 120)
# ylim(-100, 100)

subplot(122)
plot(slTime, slAltA)

show()

print "Easymini max alt (ft):",max(emAltA)
print "Stratologger max alt (ft):",max(slAltA)
```

```python
figure(figsize = (15, 5))

subplot(121)
plot(emTime, emVel)

subplot(122)
plot(slTime, slVel)

show()

print "Easymini max vel (ft/s):",max(emVel)
print "Stratologger max vel (ft/s):",max(slVel)
```

```python
slAccel = [0.0] + [dx / dt for dx, dt in zip(diff(slVel), diff(slTime))]

slAA = [x for x in slAccel] #re-make emAltA so you dont compound adjustments
for x in range(1, len(slAA)):
    if abs(slAA[x] - slAA[x-1]) > 100: #40 is minimum difference to maintain accuracy and relative "smoothness"
        slAA[x] = slAA[x-1]
        
slAAS = gaussian_filter1d(slAA, 3)
        
figure(figsize = (15, 5))

subplot(121)
plot(slTime, slAccel)

subplot(122)
plot(slTime, slAA)
plot(slTime, slAAS)

show()
```

```python
figure(figsize = (15, 5))

subplot(121)
plot(emTime, emAccel)

subplot(122)
plot(slTime, slAAS)

show()

print "Easymini max accel (ft/s^2):",max(emAccel)
print "Stratologger max accel (ft/s^2):",max(slAAS)
```

```python
maxAlt = max(slAltA)
maxVel = max(slVel)

slVel = list(slVel)
slAAS = list(slAAS)

tList = []
for x in range(0, len(slTime)): tList.append(slTime[x])
apogeeIndex = slAltA.index(maxAlt)
mainIndex = tList.index(88.5)
landIndex = tList.index(116.5)

print apogIndex
print mainIndex
print landIndex
```
